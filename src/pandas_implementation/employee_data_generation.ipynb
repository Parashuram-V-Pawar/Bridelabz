```python
import re
from faker import Faker
```

```python
fake = Faker(locale="en_IN")
n = 50
emp_data = dict()
for i in range(n):
    emp_details = dict()
    emp_id = fake.random_int(min=100000, max=999999)
    name = fake.name()
    gender = fake.random_element(elements=["Male","Female"])
    date_of_birth = fake.date_between(start_date="-65y", end_date="-22y")
    department = fake.random_element(elements= ["IT","HR","Accounts","Sales","Marketing"])
    if department == "IT":
        role = fake.random_element(elements= ["Developer","Tester","System Admin","DevOps Engineer"])
    elif department == "HR":
        role = fake.random_element(elements= ["Recruiter","HR Manager","Training Coordinator"])
    elif department == "Accounts":
        role = fake.random_element(elements= ["Accountant","Financial Analyst","Auditor"])
    elif department == "Sales":
        role = fake.random_element(elements= ["Sales Executive","Sales Manager","Account Manager"])
    else:
        role = fake.random_element(elements= ["Marketing Executive","Content Creator","SEO Specialist"])
    salary = fake.random_int(min=40000, max=90000, step=1000)
    email = f"{fake.user_name()}@gmail.com"
    phone = fake.numerify(text="##########")
    phone_number = fake.random_element(elements=['6','7','8','9']) + phone[1:]
    address = fake.address()
    address = re.sub(r'\n', ' ', address)
    address = re.sub(r',', ' ', address)

    emp_details["emp_id"] = emp_id
    emp_details["name"] = name
    emp_details["gender"] = gender
    emp_details["date_of_birth"] = date_of_birth
    emp_details["department"] = department
    emp_details["role"] = role
    emp_details["salary"] = salary
    emp_details["email"] = email
    emp_details["phone_number"] = phone_number
    emp_details["address"] = address

    emp_data[i] = emp_details

with open("employee_data.csv", "w") as f:
    f.write("Emp Id,Name,Gender,Date of Birth,Department,Role,Salary,Email,Phone,Address\n")
    for i in range(n):
        f.write(f"{emp_data[i]['emp_id']},{emp_data[i]['name']},{emp_data[i]['gender']},{emp_data[i]['date_of_birth']},{emp_data[i]['department']},"
                f"{emp_data[i]['role']},{emp_data[i]['salary']},{emp_data[i]['email']},{emp_data[i]['phone_number']},{emp_data[i]['address']}\n")
```
