```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 1. Write a Python programming to display a bar chart of the popularity of programming Languages.
data = {
    "Programming languages": ["Java", "Python", "PHP", "JavaScript", "C#", "C++"],
    "Popularity": [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    }
plt.bar(data["Programming languages"], data["Popularity"], zorder = 3)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title("Popularity of programming languages")
plt.grid(axis='y', ls= '--')
plt.show()
```

```python
# 2. Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.
plt.barh(data["Programming languages"], data["Popularity"], zorder = 3)
plt.ylabel('Programming Languages')
plt.xlabel('Popularity')
plt.title("Popularity of programming languages")
plt.grid(axis='x',ls='--')
plt.show()
```

```python
# 3. Write a Python programming to display a bar chart of the popularity of programming Languages. Use uniform color.
plt.bar(data['Programming languages'], data['Popularity'], color = 'skyblue', zorder = 3)
plt.ylabel('Programming Languages')
plt.xlabel('Popularity')
plt.title("Popularity of programming languages")
plt.grid(axis='y', ls='--')
plt.show()
```

```python
# 4. Write a Python programming to display a bar chart of the popularity of programming Languages. 
# Use different color for each bar.

bar_colors = ['red','blue','green','yellow','black','purple']
plt.bar(data['Programming languages'], data['Popularity'], color = bar_colors, zorder = 3)
plt.ylabel('Programming Languages')
plt.xlabel('Popularity')
plt.title("Popularity of programming languages")
plt.grid(axis='y', ls ='--')
plt.show()
```

```python
# 5. Write a Python programming to display a bar chart of the popularity of programming Languages. 
# Attach a text label above each bar displaying its popularity (float value).

bar_colors = ["#F48E54", "#53F3F6", "#F14262", "#5FF24C", "#7A6B4A", "#6F5ADB"]
bar = plt.bar(data['Programming languages'], data['Popularity'], color = bar_colors, zorder = 3)
plt.ylabel('Programming Languages')
plt.xlabel('Popularity')
plt.title("Popularity of programming languages")

for b in bar:
    height=b.get_height()
    plt.text(b.get_x()+b.get_width()/2,height,f"{height:.1f}",ha="center",va="bottom")

plt.grid(axis='y', ls = '--', )
plt.show()
```

```python
# 10. Write a Python program to create bar plot of scores by group and gender. Use multiple X 
# values on the same chart for men and women.
plt.figure(figsize=(8,5))
mens = [22,30,35,35,26]
womens = [25,32,30,35,29]

groups = ['G1','G2','G3','G4','G5']
bar_width = 0.5

x = np.arange(len(groups)) * (2 * bar_width)
plt.bar(x , mens, width = bar_width, label = "Men's Scores", zorder = 3)
plt.bar(x + bar_width/2, womens, width= bar_width, label = "Women's Scores", zorder = 3)

plt.legend()
plt.xticks(x, groups)
plt.grid(axis='y', ls='--')
plt.show()
```

```python
# 11. Write a Python program to create bar plot from a DataFrame.
data={
    'a':[2,4,6,8,10],
    'b':[4,2,4,2,2],
    'c':[8,3,7,6,4],
    'd':[5,4,4,4,3],
    'e':[7,2,7,8,3],
    'f':[6,6,8,6,2]
}

colors = ["#F48E54", "#53F3F6", "#F14262", "#5FF24C", "#7A6B4A", "#6F5ADB"]
bar_width = 0.7
df = pd.DataFrame(data)
df.plot(kind="bar", color = colors, edgecolor = 'black', lw=1)

plt.ylabel('Values')
plt.xlabel('Indexes')
plt.title('Plot from a DataFrame')
plt.xticks(rotation=0)
plt.show()
```

```python
# 12. Write a Python program to create bar plots with error bars on the same figure. 
data = {
   'Mean velocity': [0.2474, 0.1235, 0.1737, 0.1824],
   'Standard Deviation': [0.3314, 0.2278, 0.2836, 0.2645]
}
df = pd.DataFrame(data)
positions=range(len(df))
plt.bar(positions,df['Mean velocity'],yerr=df['Standard Deviation'],color='yellow',edgecolor='black',lw=2,capsize=5)
plt.xlabel("samples")
plt.ylabel("Mean velocity")
plt.title("Bar plots with error bars")
plt.xticks(positions, [f"Sample {i+1}" for i in positions])
plt.show()
```

```python
df=pd.DataFrame(data)
positions=range(len(df))
bars=plt.bar(positions,df['Mean velocity'],yerr=df['Standard Deviation'],color='yellow',edgecolor='black',lw=2,capsize=5)

for i ,bar in enumerate(bars):
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,height,f"{df['Mean velocity'][i]*1000}",ha='center',va='bottom')
plt.xlabel("samples")
plt.ylabel("Mean Velocity")
plt.title("Bar plots with error bars")
plt.xticks(positions, [f"Sample {i+1}" for i in positions])
plt.show()
```
