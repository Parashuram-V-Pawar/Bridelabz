```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Write a Python programming to create a pie chart of the popularity of programming Languages.
data = {
    'Programming languages': ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],
    'Popularity': [22.2, 17.6, 8.8, 8, 7.7, 6.7]
}
explodes = [0.05] * len(data['Popularity'])
plt.pie(data['Popularity'], autopct='%1.1f%%', shadow=True, explode=explodes)
plt.legend(labels = data['Programming languages'],
           title="Programming Language",
           loc = 'center left',
           bbox_to_anchor=(1.05, 0.5)
           )
plt.show()
```

```python
# 2. Write a Python programming to create a pie chart with a title of the popularity of programming Languages.
data = {
    'Programming languages': ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],
    'Popularity': [22.2, 17.6, 8.8, 8, 7.7, 6.7]
}
explodes = [0.05] * len(data['Popularity'])
plt.pie(data['Popularity'], autopct='%1.1f%%', shadow=True, explode=explodes)
plt.legend(labels = data['Programming languages'],
           title="Programming Language",
           loc = 'center left',
           bbox_to_anchor=(1.05, 0.5)
           )
plt.title("Popularity of Programming Languages", fontweight = 'bold')
plt.show()
```

```python
# 3. Write a Python programming to create a pie chart of gold medal achievements of five most successful 
# countries in 2016 Summer Olympics. Read the data from a csv file.

data = pd.read_csv('medal.csv')
explodes = [0.05] * len(data['country'])
plt.pie(data['gold_medal'], autopct='%1.1f%%', explode= explodes, shadow=True)
plt.title("Gold medal Achievement of 5 most successful countries in 2016 Summer Olympics", fontweight = 'bold')
plt.legend(labels = data['country'],
           title="Countries",
           loc = 'center left',
           bbox_to_anchor=(1.05, 0.5)
           )

plt.show()
```
